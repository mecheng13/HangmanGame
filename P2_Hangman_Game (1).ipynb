```python
# Melody Cheng
# INST126

# This is a Hangman Game with encrypted words. The player can have up to 10 incorrect guesses
# on a randomly selected, encrypted word until he or she loses. If all the letters are guessed correctly,
# the decrypted word will show.

import random
import string

wordList = ["WHDSPQZ", "XHO", "TTDBFRT", "QQJYF", "ENQD", "DNPK", "CNTR", "EHWHOD", "TSVMOHOF", "XNOCFQGTM"]

word = random.choice(wordList) # chooses a random word from the list

lives = 10
guesses = ''


#helper function that decrypts the word
def decrypt(to_decrypt):
    
    plaintext = ""
    i = 0
    
    # interating through the characters of the encrypted word and decrypts them
    while i < len(to_decrypt): 
        if i % 2 == 0:
            ascii_code = ord(to_decrypt[i]) - 1
            character = chr(ascii_code)
        else:
            ascii_code = ord(to_decrypt[i]) + 1
            character = chr(ascii_code)
            
        plaintext += character
        i += 1
        
    return plaintext


while lives > 0:   #check if player has lives left      

    # make a counter that starts with zero
    failed = 0             
  
    if(lives > 1):
        print ("You have " + str(lives) + " lives!" )
    else:
        print("You have 1 life left!")
    
    print("Letters you have already guessed: " + guesses)
    
    # for every character in secret_word    
    for char in word:  
        if char in guesses: # see if the character is in the players guess
            print (char)    # print then out the character
        else:    
            print ("-")    # if not found, print a dash
            failed += 1    # and increase the failed counter with one

    # print You Won
    if failed == 0:        
        print("You won")
        print("Your decrypted word is: ")
        print(decrypt(word))
        

    # exit the script
        break              

    
    # ask the user go guess a character
    guess_in = input("Your guess: ") 
    guess = guess_in.upper()

    # set the players guess to guesses
    guesses += guess                    

    # if the guess is not found in the secret word
    if guess not in word:  
        lives -= 1  # lives decreases with each incorrect guess      
        print ("That letter is not in the word\n")
    
    # checks if player has any lives still and notifies them if they have "died" (no lives)
    if lives == 0:  
        print("You died:( Maybe next time!")


 
   
 

 

    
    
```
